# Seaborn 
## Seaborn is the python library used for data visualization and it provides an high level interface for  drawing atractive and informative statistical graph

```
import seaborn as sns
```

```
# tips is built-in with the seaborn library
data=sns.load_dataset("tips")
data.shape
```

```
data.head()
```

```
data.corr()
```

# ploting graphs for numerical data

```
sns.heatmap(data.corr())
```

```
sns.jointplot(x="total_bill",y="tip",data=data, kind="hex")
```

```
sns.jointplot(x="total_bill",y="tip",data=data)
```

```
sns.jointplot(x="total_bill",y="tip",data=data,kind="reg")
```

```
sns.pairplot(data)
```

```
print(data["time"].value_counts())
sns.pairplot(data,hue="time")
```

```
sns.distplot(data["total_bill"])
```

```
sns.distplot(data["tip"],kde=False)
```

# Ploting graph for categorical data

```
#count ploat will give information about only one feature of data  in the dataset
print(data["time"].value_counts())
sns.countplot(x="time",data=data)
```

```
sns.countplot(x="sex",data=data)
```

```
sns.countplot(y="day",data=data)
```

```
# barplot is used to ploat the graph between two features i.e., one feature is of numeric and other is of catogorical
sns.barplot(x="sex",y="tip",data=data)
```

```
sns.boxplot("total_bill","sex",hue="day",data=data)
```

```
sns.violinplot(x="tip",y="smoker",hue="time",data=data)
```
