# use of groupby()

```python
#selects the mean of the beerservings of the particular continent
data.groupby('continent').beer_servings.mean()
```

```python
#selects the minimum wine servings of the particular continent
data.groupby('continent').wine_servings.min()
```

```python
#selects the maximum wine servings of the particular continent
data.groupby('continent').wine_servings.max()
```

```python
data.groupby('continent').continent.count()
```

# lambda function

### lambda is a function by uaing .apply we can define our own function 
### ex: lambda x:x+1,here x is input, x+1 is the output

```python
#finding the mean by using the lambda
data.groupby('continent').total.apply(lambda x:x.mean())
```

```python
#finding the range by using lambda
data.groupby('continent').total.apply(lambda x:(x.max()-x.min()))
```

```python
#finding the std using lambda
data.groupby("continent").beer_servings.apply(lambda x:x.std())
```
