```python
#2-d Scatter plot
data.plot(kind = "scatter", x= "SepalLengthCm", y = "SepalWidthCm")
plt.show()
```

```python

sns.FacetGrid(data, hue="Species" , size=4).map(plt.scatter, "SepalLengthCm" ,"SepalWidthCm" ).add_legend();
plt.show()
```

```python
x= data["SepalLengthCm"].values
y= data["SepalWidthCm"].values
z= data["PetalLengthCm"].values
```

```python
*from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x,y,z , color = "blue")
```

```python
#pair plot

sns.pairplot(data, hue = "Species" ,size =3);
plt.show()
```

```python
sns.FacetGrid(data, hue="Species" , size=5).map(sns.distplot, "PetalLengthCm").add_legend()
```

```python
data_virginica = data.loc[data['Species'] == "Iris-virginica"]

data_versicolor = data.loc[data['Species'] == "Iris-versicolor"]
data_setosa = data.loc[data['Species'] == "Iris-setosa"]
counts, bin_edges = np.histogram(data_versicolor['PetalLengthCm'], bins=10,density=True)
print(bin_edges)
print(counts)
pdf = counts/ (sum(counts))
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
plt.show()
```
